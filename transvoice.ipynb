```python
from IPython.display import Audio, display
filename = '/content/drive/MyDrive/Colab Notebooks/Audios/Simple speechs/How To Close A Door Quietly - Life Hack (128 kbps).mp3'
display(Audio(filename, autoplay=True))
```

```python
import whisper
model = whisper.load_model('tiny') #769M Parameter
```

```python
def get_audio_in_text(filename, model):
  return model.transcribe(filename)['text']
```

```python
get_audio_in_text(filename, model)
```

```python
def translate(model, filename, dest_lang):
  text = get_audio_in_text(filename, model)
  return translator.translate(text, dest=dest_lang).text
```

```python
lang = 'Portuguese' #@param ['Portuguese', 'Spanish', 'English', 'Arabic', 'Hindi', 'Bengali', 'Russian', 'Japanese', 'Lahnda']
translate(model, filename, lang)
```
